## **Série temporal**
__________________________________________________

Em [estatística](https://pt.wikipedia.org/wiki/Estat%C3%ADstica), [econometria](https://pt.wikipedia.org/wiki/Econometria), [matemática aplicada](https://pt.wikipedia.org/wiki/Matem%C3%A1tica_aplicada) e [processamento de sinais](https://pt.wikipedia.org/wiki/Processamento_de_sinais), uma **série temporal** é uma coleção de observações feitas sequencialmente ao longo do tempo. Em modelos de [regressão linear](https://pt.wikipedia.org/wiki/Regress%C3%A3o_linear) com [dados cross-section](https://pt.wikipedia.org/wiki/Dados_cross-section) a ordem das observações é irrelevante para a análise, em séries temporais a ordem dos dados é fundamental. Uma característica muito importante deste tipo de dados é que as observações vizinhas são dependentes e o interesse é analisar e modelar essa dependência.

As séries temporais existem nas mais variadas áreas de aplicação, como: [finanças](https://pt.wikipedia.org/wiki/Finan%C3%A7as), [marketing](https://pt.wikipedia.org/wiki/Marketing), [economia](https://pt.wikipedia.org/wiki/Economia), seguros, [demografia](https://pt.wikipedia.org/wiki/Demografia), ciências sociais, [meteorologia](https://pt.wikipedia.org/wiki/Meteorologia), [energia](https://pt.wikipedia.org/wiki/Energia), epidemiologia, etc.

- Fonte: https://pt.wikipedia.org/wiki/S%C3%A9rie_temporal

## **Previsão (Forecasting)**

**Previsão** (ou do inglês, _Forecasting_) é o processo de fazer previsões com base em dados passados ​​e presentes e, mais comumente, por análise de tendências. Um exemplo comum pode ser a [estimativa](https://en.wikipedia.org/wiki/Estimation) de alguma variável de interesse em alguma data futura especificada. [Predição](https://en.wikipedia.org/wiki/Prediction) (Prediction) é um termo semelhante, mas mais geral. Ambos podem referir-se a métodos estatísticos formais que empregam [séries temporais](https://en.wikipedia.org/wiki/Time_series), dados [transversais](https://en.wikipedia.org/wiki/Cross-sectional_data) ou [longitudinais](https://en.wikipedia.org/wiki/Longitudinal_study) ou, alternativamente, a métodos de julgamento menos formais. O uso pode diferir entre as áreas de aplicação: por exemplo, em [hidrologia](https://en.wikipedia.org/wiki/Hydrology), os termos "forecast" e "forecasting" são às vezes reservados para estimativas de valores em um [futuro](https://en.wikipedia.org/wiki/Future) específico vezes, enquanto o termo "prediction" é usado para estimativas mais gerais, como o número de vezes que as inundações ocorrerão durante um longo período.

[Risco](https://en.wikipedia.org/wiki/Risk) e [incerteza](https://en.wikipedia.org/wiki/Uncertainty) são centrais para _forecasting_ e _prediction_; geralmente é considerado uma boa prática indicar o grau de incerteza associado às previsões (forecasting). Em qualquer caso, os dados devem estar atualizados para que a previsão seja o mais precisa possível. Em alguns casos, os próprios dados usados ​​para prever a variável de interesse são previstos (forecast).

- Fonte: https://en.wikipedia.org/wiki/Forecasting

## **ARIMA**

Em [estatística](https://pt.wikipedia.org/wiki/Estat%C3%ADstica) e [econometria](https://pt.wikipedia.org/wiki/Econometria), particularmente em análise de [séries temporais](https://pt.wikipedia.org/wiki/S%C3%A9ries_Temporais), um __modelo auto-regressivo integrado de médias móveis__ (autoregressive integrated moving average ou __ARIMA__, na sigla em inglês) é uma generalização de um [modelo auto-regressivo de médias móveis](https://pt.wikipedia.org/wiki/ARMA) (ARMA). Ambos os modelos são ajustados aos dados da série temporal para entender melhor os dados ou para [prever](https://pt.wikipedia.org/wiki/Previs%C3%A3o_(estat%C3%ADstica)) pontos futuros na série. Modelos ARIMA são aplicados em alguns casos em que os dados mostram evidências de não estacionariedade, em que um passo inicial de diferenciação (correspondente à parte "integrada" do modelo) pode ser aplicado uma ou mais vezes para eliminar a não estacionariedade.

A parte auto-regressiva (AR) do modelo ARIMA indica que a variável evoluinte de interesse é [regressada](https://pt.wikipedia.org/wiki/Regress%C3%A3o_linear) em seus próprios valores defasados, isto é, anteriores. A parte de média móvel (MA) indica que o erro de regressão é na verdade uma [combinação linear](https://pt.wikipedia.org/wiki/Combina%C3%A7%C3%A3o_linear) dos termos de erro, cujos valores ocorreram contemporaneamente e em vários momentos no passado. A parte integrada (I) indica que os valores de dados foram substituídos com a diferença entre seus valores e os valores anteriores e este processo diferenciador pode ter sido realizado mais de uma vez. O propósito da cada uma destas características é fazer o modelo se ajustar aos dados da melhor forma possível.

Modelos ARIMA não sazonais são geralmente denotados como $ARIMA({\displaystyle p,d,q})$, em que os [parâmetros](https://pt.wikipedia.org/wiki/Par%C3%A2metro_(estat%C3%ADstica)) ${\displaystyle p, d}$ e ${\displaystyle q}$ são números inteiros não negativos, ${\displaystyle p}$ é a ordem (número de defasagens) do modelo auto-regressivo, ${\displaystyle d}$ é o grau de diferenciação (o número de vezes em que os dados tiveram valores passados subtraídos) e ${\displaystyle q}$ é a ordem do modelo de média móvel. Modelos ARIMA sazonais são geralmente denotados como $ARIMA({\displaystyle p,d,q})({\displaystyle P,D,Q}){\displaystyle _{m}}$, em que ${\displaystyle m}$ se refere ao número de períodos em cada temporada e ${\displaystyle P}$, ${\displaystyle D}$ e ${\displaystyle Q}$ se referem aos termos de auto-regressão, diferenciação e média móvel para a parte sazonal do modelo ARIMA.

- Fonte: https://pt.wikipedia.org/wiki/ARIMA

```python
# função para decompor a série temporal
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# decompondo a série pra avaliar tendência, sazonalidade e resíduos
result = seasonal_decompose(df)
result.plot();
```

```python
# plotando gráficos maiores
fig, (ax1,ax2,ax3, ax4) = plt.subplots(4,1, figsize=(16,8))

result.observed.plot(ax=ax1)
result.trend.plot(ax=ax2)
result.seasonal.plot(ax=ax3)
result.resid.plot(ax=ax4)
plt.tight_layout();
```

### **Teste de estacionariedade**

**Teste de estacionariedade:**

- A hipótese nula é que a série não é estacionária, ou seja, se o p-valor for menor que 0,05, rejeitamos que a série não é estacionária. 
- Caso seja maior, não podemos descartar que a série não é estacionária.


```python
# função para teste ADF
from statsmodels.tsa.stattools import adfuller
```

```python
# visualizando o resultado
result = adfuller(df['production'].dropna())

print(
    f'Teste ADF: {result[0]}',
    f'p-value: {result[1]}',
    sep='\n'
)
```

>  A série não é realmente estacionária (p-valor > 0.05, não podemos rejeitar a hipótese nula).

```python
# função ".diff()", calcula a diferença de um elemento do dataframe em comparação 
# com outro elemento no dataframe (o padrão é o elemento da linha anterior)
result = adfuller(df['production'].diff().dropna())

print(
    f'Teste ADF: {result[0]}',
    f'p-value: {result[1]}',
    sep='\n'
)
```

```python
# auto ARIMA nos ajuda a buscar os melhores parâmetros
fit_arima = auto_arima(df, d=1, start_p=1, start_q=1, max_p=3, max_q=3, 
                       seasonal=True,
                       n=6,
                       D=1,
                       start_P=1,
                       start_Q=1,
                       max_P=2,
                       max_Q=2,
                       information_criterion='aic',
                       trace=True,
                       error_action='ignore',
                       stepwise=True)
```

> Agora que sabemos os melhores parâmetros, vamos criar um modelo SARIMAX (versão do arima com parametros para sazonalidade).

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
model = SARIMAX(df, order=(1,1,3), 
                seasonal_order=(0,0,0,0))

resultado_sarimax = model.fit()
```

```python
print(resultado_sarimax.summary())
```

```python
# realizando um "backtesting" com predições de valores dos 36 meses 
preds = resultado_sarimax.get_prediction(start=-36)
pred_mean = preds.predicted_mean

pred_mean
```

```python
# intervalo de confiança
intervalo_conf = preds.conf_int()

limite_inferior = intervalo_conf.iloc[:, 0]
limite_superior = intervalo_conf.iloc[:, 1]
```

```python
limite_inferior[0], limite_superior[0]
```

```python
pred_mean[0]
```

```python
datas_preds = np.asarray(pred_mean.index)
datas = np.asarray(df.index)
```

```python
plt.figure(figsize=(16,8))


plt.plot(datas_preds, pred_mean.values, 
         color='red', label='prediction')

plt.fill_between(datas_preds, 
                 limite_inferior, 
                 limite_superior,
                 color='red')


plt.plot(datas, df.values, label='real')
plt.legend()
plt.show();
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
rmse_sarima = sqrt(mean_squared_error(df[-36:].values, 
                                     pred_mean.values))

rmse_sarima
```
